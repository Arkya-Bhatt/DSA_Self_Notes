# ARRAYS

```python
A = [1,2,3]
print(A)
```

```python
# Append - Insert element at the end of an array - On average: O(1)
A.append(5)

print(A)
```

```python
# Pop - Deleting element at end of array - O(1)
A.pop()

print(A)
```

```python
# Insert (not at end of array) - O(n)
A.insert(2, 5)

print(A)
```

```python
# Modify an element - O(1)
A[0] = 7

print(A)
```

```python
# Accessing element given index i - O(1)
print(A[2])
```

```python
# Checking if array has an element - O(n)
if 7 in A:
    print(True)
```

```python
# Checking length - O(1)
print(len(A))
```

# STRINGS

```python
# Append to end of string - O(n)
s = 'hello'

b = s + 'z'

print(b)
```

```python
# Check if something is in string - O(n)
if 'f' in s:
    print(True)
```

```python
# Check length of string - O(1)
len(s)
```
