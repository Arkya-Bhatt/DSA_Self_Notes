# STACKS - LAST IN FIRST OUT (LIFO)

```python
stk = []
print(stk)
```

```python
# Append to top of stack - O(1)
stk.append(5)
```

```python
stk.append(4)
stk.append(3)

stk
```

```python
# Pop from stack - O(1)
x = stk.pop()

print(x)
print(stk)
```

```python
# Ask what's on the top of stack - O(1)
# Peek Operation on Stack - O(1)
print(stk[-1])
```

```python
# Ask if something is in the stack - O(1)
# isEmpty() or isFull() Operation on Stack - O(1)
if stk:
    print(True)
```

# QUEUES - FIRST IN FIRST OUT (FIFO)

# DEQUE

```python
from collections import deque

q = deque()

print(q)
```

```python
# Enqueue - Add element to the right - O(1)
q.append(5)
q.append(6)
q.append(7)
q
```

```python
# Deqeue (pop left) - Remove element from the left - O(1)
q.popleft()
```

```python
q
```

```python
# Peek from left side - O(1)
q[0]
```

```python
# Peek from right side - O(1)
q[-1]
```
